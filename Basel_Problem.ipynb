```python
import math
import matplotlib.pyplot as plt
import numpy as np

def summ(n):
    sum1 = 0 
    for i in range(1,n+1):
        sum1 += 1/i**2 
    return(sum1)

def basel_calc(n):
    array_x = list((range(1,n+1)))
    array_y = []

    for i in range(1,n+1):
        array_y.append(summ(i))
    
    x_points = np.array(array_x)
    y_points = np.array(array_y)

    plt.axhline(y= math.pi*math.pi/6, xmin=0, xmax=10000000, color='r', linestyle='-.', linewidth=1.5)
    plt.plot(x_points,y_points)
    plt.xlabel("n", size =20)
    plt.ylabel("Summation 1/n^2" , size =15)
    plt.show()
```

```python
def percent_error_pi(n):
    true_value = math.pi
    sum1 = summ(n)
    calc_value = (6*sum1)**0.5
    
    error = (true_value - calc_value)/true_value
    p_error = error*100
    
    return(p_error)
```

```python
def percent_error_graph(n):
    array_x = list((range(1,n+1)))
    array_y = []
    for i in range(1,n+1):
        array_y.append(percent_error_pi(i))
    
    x_points = np.array(array_x)
    y_points = np.array(array_y)
    
    plt.plot(x_points,y_points)
    plt.axhline(y= 0, xmin=0, xmax=10000000, color='r', linestyle='-.', linewidth=1.5)
    plt.xlabel("n", size =20)
    plt.ylabel("Percentage Error" , size =15)
    plt.show()
    
```

```python
basel_calc(10)
```

```python
basel_calc(100)
```

```python
basel_calc(1000)
```
